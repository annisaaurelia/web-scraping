```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import re
import pandas as pd
import datetime
import time
import math
```

```python
driver = webdriver.Firefox(executable_path='C:/Users/Annisa/Documents/Scraping/scraping_script_final/scraping_script_final/geckodriver.exe')
link = "https://www.rumah123.com/jual/dki-jakarta/apartemen/?minPrice=300000000&maxPrice=400000000&transactedIncluded=1&sort=posted-desc&"
result = pd.DataFrame()

for page in range(1,101):
    driver.get(link+'page='+str(page))
    item = driver.find_elements_by_class_name('sc-eCssSg')
    print("page: " + str(page))

    for x in range(0,20):
        # get url
        try:
            page_detail_link = (item[x].find_elements_by_class_name('ui-atomic-link')[0].get_attribute("href"))
            # print("page: " + str(page) + " - item: " + str(x)+" - " + page_detail_link)

            # Open a new window
            driver.execute_script("window.open('');")
            # Switch to the new window
            driver.switch_to.window(driver.window_handles[1])
            driver.get(page_detail_link)
            time.sleep(1)

            # get attribute information
            title = driver.find_element_by_class_name('ui-organisms-listing-description-r123__detail').find_element_by_class_name("ui-atomic-text").text
            price = re.sub("\.","",re.sub("Rp. ","",driver.find_elements_by_class_name('r123-o-listing-summary__detail--price')[0].text))
            area = driver.find_element_by_class_name('r123-o-listing-summary__detail--address').text
            luas_bangunan = re.sub("Luas Bangunan : | mÂ²","",driver.find_element_by_class_name('r123-o-listing-summary__detail--area-info').find_element_by_class_name("ui-molecules-list__item").text)
            status = driver.find_element_by_class_name("r123-o-listing-summary__flagging-badge").text

            try:
                kamar_tidur = driver.find_element_by_class_name('r123-o-listing-summary__detail--attribute').find_elements_by_class_name("ui-molecules-list__item")[0].text
                kamar_mandi = driver.find_element_by_class_name('r123-o-listing-summary__detail--attribute').find_elements_by_class_name("ui-molecules-list__item")[1].text
            except:
                kamar_tidur = "-"
                kamar_mandi = "-"

            num_detail = len(driver.find_element_by_class_name('ui-molecules-tab-list-r123-detail').find_elements_by_class_name("ui-atomic-text"))
            detail = driver.find_element_by_class_name('ui-molecules-tab-list-r123-detail').find_elements_by_class_name("ui-atomic-text")
            y = 0
            id_ads = ""
            furnished = ""
            jumlah_lantai = 0
            tanggal_tayang = ""
            sertifikat = ""

            while(y < num_detail):
                if(detail[y].text == "ID Iklan"):
                    id_ads = detail[y+1].text
                elif(detail[y].text == "Dilengkapi Perabotan"):
                    furnished = detail[y+1].text
                elif(detail[y].text == "Jumlah Lantai"):
                    jumlah_lantai = detail[y+1].text
                elif(detail[y].text == "Tanggal Tayang"):
                    tanggal_tayang = detail[y+1].text
                elif(detail[y].text == "Sertifikat"):
                    sertifikat = detail[y+1].text
                y+=2

            # Return result
            date = datetime.datetime.now().strftime("%Y-%m-%d")
            new_result = pd.DataFrame([{'title':title,'price':price, 'id_ads':id_ads, 'furnished':furnished, 'jumlah_lantai':jumlah_lantai,'tanggal_tayang':tanggal_tayang,'sertifikat':sertifikat,
                                       'area':area, 'luas_bangunan':luas_bangunan, 'kamar_tidur':kamar_tidur, 'kamar_mandi':kamar_mandi, 'status':status,'link':page_detail_link,'scraping_date':date}])
            result = result.append(new_result)

        except:
            pass

        # close the active tab
        driver.close()
        driver.switch_to.window(driver.window_handles[0])
    
    result.to_csv("result/Apartment DKI Jakarta 300 - 400 jt.csv",index=False)
```
